## Reading Dataset

```python
#Read data
data_folder = 'data/'
data_1_4 = 'data_supervised_1_of_4.json'
data_2_4 = 'data_supervised_2_of_4.json'
data_3_4 = 'data_supervised_3_of_4.json'
data_4_4 = 'data_supervised_4_of_4.json'
with open (data_folder+data_1_4) as json_file:
    data_1 = pd.DataFrame(json.load(json_file))
with open (data_folder+data_2_4) as json_file:
    data_2 = pd.DataFrame(json.load(json_file))
with open (data_folder+data_3_4) as json_file:
    data_3 = pd.DataFrame(json.load(json_file))
with open (data_folder+data_4_4) as json_file:
    data_4 = pd.DataFrame(json.load(json_file))
```

## 2) TextRank BM25 Predictions

Based in Mahalsea and Tarau (2000) Text Rank Algorithm. But using the BM25 similarity function (Barrios et al., 2016). 

In order to use TextRank alg as provided by gensim, the leads processed are the ones with more than 10 sentences. For the leads with length less or equal to 10 sentences, is assumed that the summary is equivalent to the document.

```python
text_rank_pred = []
for i in range (0, len(data)):
    try:
        if (data['nb_sentences_doc'][i] > 10):
            text_rank_pred.append(summarize(data['clean_lead'][i], split= True))
        else:
            text_rank_pred.append(data['lead_sentences'][i])
    except:
        text_rank_pred.append('')
        print('Error at lead ',i)
    if (i%100 == 0):
        print(i, ' leads processed out of ',len(data))
```

```python
summaries = []
for summary in text_rank_pred:
    try:
        summaries.append(sentences_to_summary(summary))
    except: 
        summaries.append('')
        print('Error at lead ',text_rank_pred.index(summary))
```

```python
with open('data/text_rank_bm25_summaries.json', 'w') as f:
    json.dump(summaries, f)
```

```python
def vector_to_text_files(vector, folder_path, name):
    for i in range(0,len(vector)):
        with open(folder_path+name+'.'+str(i)+'.txt', 'w', encoding='utf-8') as f:
            json.dump(vector[i], f, ensure_ascii=False, indent=4)
    return 'Text files exported!'
```
